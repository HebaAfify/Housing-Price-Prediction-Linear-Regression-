# Model training and Evaluation

```python
x = df.drop(['median_house_value'], axis=1)
y = df['median_house_value']
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2)
```

```python
training_data = train_x.join(train_y)
```

```python
testing_data = test_x.join(test_y)
```

```python
train_x.shape
```

# 1 - Training by XGboost

```python
import xgboost as xgb 
```

```python
from xgboost import XGBRegressor
```

```python
train_x, train_y = training_data.drop(['median_house_value'], axis=1), training_data['median_house_value']

xgb_df = xgb.XGBRegressor(objective ='reg:squarederror',
                  n_estimators = 10, seed = 123)

xgb_df.fit(train_x, train_y)
```

# Model Evaluation 

```python
from sklearn.metrics import mean_squared_error as MSE

pred = xgb_df.predict(test_x)

rmse = np.sqrt(MSE(test_y, pred))
print("RMSE : % f" %(rmse))
```

```python
plt.hist(test_y - pred)
```

```python
plt.scatter(test_y, pred)
```
